```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Looking Into the Dataset

```python
karate_club = nx.karate_club_graph()

print("Number of nodes:", karate_club.number_of_nodes())
print("Number of edges:", karate_club.number_of_edges())
```

```python
plt.figure(figsize=(8, 6))
nx.draw(karate_club, with_labels=True, node_color='skyblue', node_size=500, edge_color='gray')
plt.title("Zachary's Karate Club Network")
plt.show()
```

A simple look in analyzing the Zachary's Karate Club dataset, it is evident that the network exhibits a clear division into two distinct communities or groups. This observation suggests the presence of cohesive subgroups within the karate club, reflecting underlying social dynamics and relationships among its members. 

### Computing Network Measures

```python
# Compute degree centrality
degree_centrality = nx.degree_centrality(karate_club)
print("Degree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"Node {node}: {centrality}")

# Compute betweenness centrality
betweenness_centrality = nx.betweenness_centrality(karate_club)
print("\nBetweenness Centrality:")
for node, centrality in betweenness_centrality.items():
    print(f"Node {node}: {centrality}")

# Compute clustering coefficient
clustering_coefficient = nx.clustering(karate_club)
print("\nClustering Coefficient:")
for node, coefficient in clustering_coefficient.items():
    print(f"Node {node}: {coefficient}")
```

After looking at the numbers, it seems like some people in the karate club know a lot of others (degree centrality), while others seem to act as bridges between different groups (betweenness centrality). Also, there are some tight-knit groups where everyone knows each other well (clustering coefficient)

### Analyzing Community Structure

Applying community detection algorithms to identify distinct communities or subgroups within the karate club network. Visuzalizing the detected communities to better understand the network's community structure.


```python
# Apply Girvan-Newman algorithm for community detection
from networkx.algorithms.community import girvan_newman

# Detect communities
communities_generator = girvan_newman(karate_club)
communities = tuple(sorted(c) for c in next(communities_generator))

# Visualize community structure
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(karate_club)  # Layout for visualization
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']  # List of colors for visualizing communities
for i, community in enumerate(communities):
    nx.draw_networkx_nodes(karate_club, pos, nodelist=community, node_color=colors[i], node_size=300)
nx.draw_networkx_edges(karate_club, pos, alpha=0.5)
plt.title("Community Structure of Zachary's Karate Club (Girvan-Newman)")
plt.axis('off')
plt.show()
```

The application of community detection algorithms to the Zachary's Karate Club dataset has revealed the presence of two distinct groups within the network. This separation into two cohesive communities reflects the underlying social dynamics and division within the karate club, shedding light on the group affiliations and interactions among its members. Such insights offer valuable understanding of the community structure and relationships within the club, contributing to a deeper appreciation of its organizational dynamics. Identifying two distinct groups within the Zachary's Karate Club network highlights the club's social divisions, providing valuable insights into its internal dynamics and member interactions.

The Girvan-Newman algorithm was utilized for community detection in the karate club network. This algorithm gradually identifies distinct groups by iteratively removing 'bridge' connections between nodes, resulting in the observed separation of the network into two cohesive communities.

### K-Means CLustering Technique

```python
from sklearn.cluster import KMeans

G = nx.karate_club_graph()

# Create a feature matrix based on node degrees
X = np.array([list(dict(G.degree).values())]).T

# Perform K-means clustering
kmeans = KMeans(n_clusters=2)  # Assuming we want to divide the club into 2 clusters
kmeans.fit(X)

# Get the labels assigned by K-means
labels = kmeans.labels_

# Print node degrees and cluster labels
for node, degree in dict(G.degree).items():
    print(f"Node {node}: Degree = {degree}, Cluster = {labels[node]}")

# Draw the graph with nodes colored by their cluster
pos = nx.spring_layout(G)  # Position nodes using Fruchterman-Reingold force-directed algorithm
nx.draw(G, pos, node_color=labels, cmap=plt.cm.Set1, with_labels=True)
plt.title("K-means Clustering on Zachary's Karate Club")
plt.show()
```

Overall Interpretation:
- The K-means algorithm has divided the Zachary's Karate Club network into two clusters based on the degree centrality of nodes.
- Cluster 0 contains nodes with lower degrees and is likely composed of peripheral or less central members of the club.
- Cluster 1 contains nodes with higher degrees and likely represents more central or influential members of the club.
- The clustering provides insight into the network structure and the roles of individual nodes within the club.
